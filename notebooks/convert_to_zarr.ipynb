# Convert to zarr

```python
from src.muvis_align.MVSRegistration import MVSRegistration
```

## Initialise parameters

```python
operation = {
	'operation': 'convert',
	'input':
	{
		'path': '../data/*/*.tiff',
		'source_metadata':
		{
			'scale': {'x': 0.032, 'y': 0.032},
			'position': {'y': 'fn[-3]*24', 'x': 'fn[-2]*24'}
		}
	},
	'output': '../../output/'
}
```

## Load data

```python
reg = MVSRegistration()
reg.init_params({}, operation)
sims = reg.init_sims()
```

## Store as zarr

```python
for label, sim in zip(reg.file_labels, sims):
    print(f'Writing {label}')
    reg.save(label, sim)
```
