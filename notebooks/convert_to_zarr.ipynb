# Convert to zarr

```python
from src.muvis_align.MVSRegistration import MVSRegistration
from src.muvis_align.util import dir_regex
```

## Initialise parameters

```python
operation = {
	'operation': 'convert',
	'input': '../data/*/*.tiff',
	'source_metadata':
	{
		'scale': {'x': 0.032, 'y': 0.032},
		'position': {'y': 'fn[-3]*24', 'x': 'fn[-2]*24'}
	},
	'output': '../../output/'
}
```

## Get filenames

```python
filenames = dir_regex(operation['input'])
filenames
```

## Load data

```python
reg = MVSRegistration({'output': {'ome_version': '0.5'}})
reg.init('', filenames, operation)
sims = reg.init_sims()
```

## Store as zarr

```python
for label, sim in zip(reg.file_labels, sims):
    reg.save(reg.output + label, sim)
```
